```python
%matplotlib inline
import pandas as pd
import os,sys
import numpy as np
import math 
import matplotlib.pyplot as plt

np.set_printoptions(precision=8, suppress=True)
```

```python
fl = open(r"KhAB/n1.txt", "r")
u1 = np.zeros((101, 4))
i = 0
for line in fl:
    tmp = line.split(" ")
    u1[i][0] = float(tmp[0])
    u1[i][1] = float(tmp[1])
    u1[i][2] = float(tmp[2])
    u1[i][3] = float(tmp[3])
    i += 1
fl.close()
print(u1)
```

```python
fl = open(r"Bakh/n2.txt", "r")
u2 = np.zeros((101, 4))
i = 0
for line in fl:
    tmp = line.split(" ")
    u2[i][0] = float(tmp[0])
    u2[i][1] = float(tmp[1])
    u2[i][2] = float(tmp[2])
    u2[i][3] = float(tmp[3])
    i += 1
fl.close()
print(u2)
```

```python
d = abs(u1 - u2)/u1*100
dl = np.zeros(len(u1))
for i in range(len(dl)):
    dl[i] = max(d[i])
print(max(dl))    
print(d)    
```
